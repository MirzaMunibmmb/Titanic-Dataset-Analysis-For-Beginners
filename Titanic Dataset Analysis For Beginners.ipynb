# Does survival have anything to do with the class in which the passenger was traveling and who survived more 1st class passengers, 2nd class passengers or 3rd class passengers?

### One Way of Solving Question

```python
print("*"*50)

p1_class = (titanic_data["Pclass"] == 1 ) &  (titanic_data["Survived"] == 1 ) #getting  Survived passengers in P1 Class

print("Total Passengers Survived in P1 Class In Titanic were: ", p1_class.sum())  #getting number of Survived passengers in P1 Class

p2_class = (titanic_data["Pclass"] == 2 ) &  (titanic_data["Survived"] == 1 )  #getting  Survived passengers in P2 Class

print("Total Passengers Survived in P2 Class In Titanic were: ", p2_class.sum())  #getting number of Survived passengers in P2 Class


p3_class = (titanic_data["Pclass"] == 3 ) &  (titanic_data["Survived"] == 1 )  #getting Survived passengers in P3 Class

print("Total Passengers Survived in P3 Class In Titanic were: ", p3_class.sum())  #getting number of Survived passengers in P3 Class

print("*"*50)

if (p1_class.sum() > p2_class.sum() ) & ( p1_class.sum() > p3_class.sum() ) :  #Checking if P1 survived most
    print("P1 Class Survived the Most", p1_class.sum())
    
elif (p2_class.sum() > p1_class.sum() ) & ( p2_class.sum() > p3_class.sum() ) : #Checking if P2 survived most
    print("P2 Class Survived the Most", p2_class.sum())
    
else:
    print("P3 Class Survived the Most", p3_class.sum()) #Checking if P3 survived most

print("*"*50)
```

### Second Way of Solving Question

```python
print("*"*50)
passenger_class = titanic_data.groupby(["Pclass"])["Survived"].sum() #getting  Survived passengers in PClass

print(passenger_class)

l = max(passenger_class) #getting  which Pclass Survived most.

print("*"*50)

print("P1 Class Survived most:", l)

print("*"*50)
```

# Which age group, survived the most, kids (0-12), teen agers (13-19), youngsters (20-30), middle age people (31-50) or elders (50 and) greater) ?

```python
x = []

print("*"*50)

kids = (titanic_data["Age"] <= 12 ) &  (titanic_data["Survived"] == 1 )  #getting  Survived passengers from Age 0 to 12

print("Total Kids Survived In Titanic were: ", kids.sum()) #getting sum
w
teenagers = (titanic_data["Age"] > 12 ) & (titanic_data["Age"] <= 19 ) &  (titanic_data["Survived"] == 1 ) #getting  Survived passengers from Age 13 to 19

print("Total Teengers Survived In Titanic were: ", teenagers.sum()) #getting sum

youngsters = (titanic_data["Age"] > 19 ) & (titanic_data["Age"] <= 30 ) &  (titanic_data["Survived"] == 1 ) #getting  Survived passengers from Age 20 to 30

print("Total Youngers Survived In Titanic were: ", youngsters.sum()) #getting sum

mid_age_people = (titanic_data["Age"] > 30 ) & (titanic_data["Age"] <= 50 ) &  (titanic_data["Survived"] == 1 ) #getting  Survived passengers from Age 31 to 50

print("Total Mid Age People Survived In Titanic were: ", mid_age_people.sum()) #getting sum

elders =  (titanic_data["Age"] > 50 ) &  (titanic_data["Survived"] == 1 ) #getting  Survived passengers older than 50

print("Total Elders Survived In Titanic were: ", elders.sum()) #getting sum

print("*"*50)

x.append(kids.sum())
x.append(teenagers.sum())
x.append(youngsters.sum())
x.append(mid_age_people.sum())
x.append(elders.sum())

Maximum = max(x)

print("The Most Survived Age group Of People are Youngsters:" , Maximum) #getting Which age group survived most

print("*"*50)
```

# Which people survived the most, either those who were traveling with the family or those who were traveling alone?

```python
print("*"*50)

family = titanic_data.groupby(["SibSp"])["Survived"].sum()  #checkingg who survived most ipe Single or with family members

print(family)

print("*"*50)

y = max(family)  #getting either Family people survived or Single people survived most

print("The People who were traveling alone Survived the most:", y)

print("*"*50)
```

```python
print("*"*50)

single = (titanic_data["SibSp"] == 0 ) &  (titanic_data["Survived"] == 1 ) #getting  Single Survived passengers

print("Total Single People Survived In Titanic were: ", single.sum())

print("*"*50)

with_family = (titanic_data["SibSp"] != 0 ) &  (titanic_data["Survived"] == 1 ) #getting  Family members Survived 

print("Total Family People Survived In Titanic were: ", with_family.sum())

print("*"*50)


if single.sum() > with_family.sum() : #checking if number of Single were higher than family
    print("Single People Survived the Most", single.sum()) #getting  sum of Single people
    
elif single.sum() < with_family.sum(): #checking if number of Family were higher than Single
    print("Family People Survived the Most", with_family.sum()) #getting  sum of family people

```
